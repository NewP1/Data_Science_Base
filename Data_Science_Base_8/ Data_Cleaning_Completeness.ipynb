# 결측값 : 비어있는 값

## 결측값이 있다는 것은 데이터가 완결성이 없는 것 => 데이터 분석에 방해가 될 수 있다.
            
## 결론적으로는 판단하기가 어렵다는 뜻.

### 결측값이 없는게 좋지만 자주 발생한다면 이유를 파악해야한다. 

```python
import pandas as pd
```

```python
df = pd.read_csv('data/attendance.csv', index_col=0)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()  # 이렇게 쉽게 결측값을 찾을 수 있다.
```

### 1. 결측값을 그냥 둔다.
### 2. 결측값이 있는 레코드를 아예 지워버린다.
### 3. 특정 column을 포기하는 방법.
### 4. 있는 데이터를 모두 보존하는 방법.

```python
df.dropna() # 기존 데이터 프레임을 건들이지 않고 새로운 데이터프레임을 만드는 것이다. 행을 삭제.
```

```python
df.dropna(inplace=True) # 이렇게 하면 기존 데이터 프레임을 변경
df
```

```python
df = pd.read_csv('data/attendance.csv', index_col=0)
df
```

```python
df.dropna(axis='columns') # 결측값이 있는 column을 삭제( 배구 )
```

```python
df.fillna(0) # 결측값이 있던 부분을 0으로 대체해서 기존 데이터들을 보존
```

```python
df.mean()
```

```python
df.fillna(df.mean()) # 결측값을 그 column의 평균값으로 대체
```

```python
df.median()
```

```python
df.fillna(df.median()) # 결측값을 중앙값으로 대체
```

```python
df
```

```python
df.fillna(df.median(), inplace=True) # 기존 데이터프레임에 적용
df
```
