```python
import pandas as pd
import seaborn as sns
```

```python
laptops_df = pd.read_csv('data/laptops.csv')
laptops_df.head()
```

```python
laptops_df['os'].unique()
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='box')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='violin')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip')   # kind = strip
```

```python
laptops_df['processor_brand'].unique()
```

### hue 파라미터 : 색을 의미, 주어진 파라미터에 따라 색을 달리 해서 보여줌 👇

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip', hue='processor_brand')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='swarm', hue='processor_brand')   # kind = swarm 데이터들이 겹치지 않고 펼쳐짐
```
