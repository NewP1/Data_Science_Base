# 클러스터 분석 (Cluster Analysis)
* 데이터들을 몇 가지 부류로 나누어 분석하는 것  


```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
```

```python
df.head()
```

```python
interests = df.loc[:,'History':'Pets']
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr)
```
