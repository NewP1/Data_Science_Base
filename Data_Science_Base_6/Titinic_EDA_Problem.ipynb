* 생존 여부는 'Survived' column에 저장되어 있습니다. 0이 사망, 1이 생존을 의미합니다.
* 좌석 등급은 'Pclass' column에 저장되어 있습니다. 1은 1등실, 2는 2등실, 3은 3등실을 의미합니다.
* 지불한 요금은 'Fare' column에 저장되어 있습니다.

1. 타이타닉의 승객은 30대와 40대가 가장 많다.
2. 가장 높은 요금을 낸 사람은 30대이다.  ✔✔✔
3. 생존자가 사망자보다 더 많다.
4. 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다.  ✔✔✔
5. 가장 생존율이 높은 객실 등급은 1등실이다.  ✔✔✔
6. 나이가 어릴수록 생존율이 높다.

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/titanic.csv', index_col=0)
df.head()
```

```python
sns.violinplot(data=df, y='Age')
```

```python
df.plot(kind='hist', y='Age', bins=30)
```

```python
sns.jointplot(data=df, x='Age', y='Fare')
```

```python
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df.plot(kind='scatter', x='Pclass', y='Survived')
```

```python
sns.kdeplot(df['Pclass'], df['Survived'])  ##보기 5: "가장 생존율이 높은 객실 등급은 1등실이다."
```

```python
# 생존 여부에 따른 나이 분포

sns.stripplot(data=df, x="Survived", y="Age")
```

```python
sns.violinplot(data=df, x='Survived', y='Age')
```

```python
sns.stripplot(data=df, x='Survived', y='Age', hue='Sex')
```
