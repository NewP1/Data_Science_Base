## 합치는 4가지 방법
1. inner join
 - A와 B에서 겹치는 부분만 합치는 것(교집합)
2. left outer join
 - A와 B 중 A에 있는 것을 기준으로 합치는 것
3. right outer join
 - A와 B 중 B에 있는 것을 기준으로 합치는 것
4. full outer join
 - A와 B에 있는 모든 것들을 합치는 것(합집합)

```python
import pandas as pd
```

```python
df_p = pd.read_csv('data/vegetable_price.csv')
df_q = pd.read_csv('data/vegetable_quantity.csv')
```

```python
df_p
```

```python
df_q
```

```python
pd.merge(df_p, df_q, on='Product') # inner join
```

```python
pd.merge(df_p, df_q, on='Product', how='left') # left outer join
```

```python
pd.merge(df_p, df_q, on='Product', how='right') # right outer join
```

```python
pd.merge(df_p, df_q, on='Product', how='outer') # full outer join
```
